# Multiple regression is like linear regression, but with more than one independent value, meaning that we try to predict a value based on two or more variables.

```python
import pandas
df=pandas.read_csv("data.csv")
df
```

```python
from sklearn.linear_model import LinearRegression

X = df[['Weight', 'Volume']]
y = df[['CO2']]

regr = LinearRegression()
regr.fit(X, y)
```

```python
#Now predict Co2 value based on cars weight and volume
#predict the CO2 emission of a car where tche weight is 2300kg, and the volume is 1300cm3:
predictC02=regr.predict([[2300,1300]])
predictC02
```

#Coefficient


```python
print(regr.coef_)
```

The result array represents the coefficient values of weight and volume.

Weight: 0.00755095
Volume: 0.00780526

These values tell us that if the weight increase by 1kg, the CO2 emission increases by 0.00755095g.

And if the engine size (Volume) increases by 1cm3, the CO2 emission increases by 0.00780526g.

```python
print(regr.intercept_)
```

```python
print("Coefficients:", regr.coef_)
```

# predicted_CO2 = intercept + (weight_coef * weight) + (volume_coef * volume) We have already predicted that if a car with a 1300cm3 engine weighs 2300kg, the CO2 emission will be approximately 107g.

What if we increase the weight with 1000kg?

107.2087328 + (1000 * 0.00755095) = 114.75968
