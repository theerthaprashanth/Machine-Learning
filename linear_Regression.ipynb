```python
from sklearn import linear_model

reg = linear_model.LinearRegression()   # ← capital “L” and “R”
reg.fit(df[['area']], df['price'])

```

```python
reg.predict([[3300]])
```

# Canadas Income Prediction in the year of 2020


```python
k=pd.read_csv('canada_per_capita_income.csv')
k
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(k['year'],k['per capita income (US$)'],color='pink',marker='*')
plt.show()
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()  
reg.fit(k[['year']], k['per capita income (US$)'])
# Trains the linear regression model 'reg' using 'year' as input (X) and 'per capita income (US$)' as output (y)
# The model will learn the best-fitting line to predict income from year
```

```python
reg.predict([[2020]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y=mx+c
828.465*2020+(-1632210)
```
