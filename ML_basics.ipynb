# lINEAR REGRESSION

```python
#he main goal is to find the relationship between input and output.

#Once the relationship is known, you can predict unknown outputs.

#Its used when the output is a number, not a category.
#Q. Predict the age of 10 year's old car
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

speed = myfunc(10)

print(speed)
```

```python
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
```

#Independent variable = What YOU choose (what to order)
#Dependent variable = What you GET (the bill amount)
#Smaller std_err  slope estimate is more precise
#Larger std_err  slope estimate is less precise




slope, intercept, r, p, std_err = stats.linregress(x, y) is from **linear regression** in Python using `scipy.stats`. Linear regression is just a way to find the **best straight line** that fits your data points `(x, y)`.

1. **slope**  How steep the line is.

   * If slope = 2, it means for every 1 unit increase in `x`, `y` increases by 2 units.
   * Positive slope  line goes up, Negative slope  line goes down.

2. **intercept**  Where the line crosses the y-axis.

   * Think of it as the starting value of `y` when `x = 0`.

3. **r**  Correlation coefficient.   if there is no relationship we can't use this prediction,therefore , r, should hold some value

   * Measures how strongly `x` and `y` are related.
   * Value between -1 and 1.

     * 1  perfectly positive relationship,
     * -1  perfectly negative,
     * 0  no relationship.

4. **p**  p-value.

   * Tells if the relationship is statistically significant.
   * Small p (like < 0.05)  relationship is probably real, not by chance.

5. **std_err**  Standard error of the slope.

   * Tells how much the slope could vary if you used a different dataset.
   * Smaller value  slope is more reliable.

```python
#These values for the x- and y-axis should result in a very bad fit for linear regression:
import matplotlib.pyplot as plt
from scipy import stats

x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```
