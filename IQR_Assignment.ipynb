```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
matplotlib.rcParams['figure.figsize']=(10,6)
```

```python
df=pd.read_csv('height_weight.csv')
df
```

```python
plt.hist(df.height,bins=28,rwidth=0.8)
plt.xlabel('Height(inches)')
plt.ylabel('count')
plt.show()
```

```python
plt.hist(df.weight,bins=28,rwidth=0.8)
plt.xlabel('Weight(inches)')
plt.ylabel('count')
plt.show()
```

# Detect Outliers Based on Weight

```python
Q1=df.height.quantile(0.25)
Q3=df.height.quantile(0.75)
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit,upper_limit
```

```python
df_outlier=df[(df.height<lower_limit)|(df.height>upper_limit)]
df_outlier
```
