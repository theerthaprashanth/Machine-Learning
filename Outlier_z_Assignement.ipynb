```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
matplotlib.rcParams['figure.figsize']=(10,6)
```

```python
df=pd.read_csv('bhp.csv')
df
```

# Remove Outlier Using Percentile Technique

```python
min_threshold,max_threshold=df.price_per_sqft.quantile([0.001,0.999])
min_threshold,max_threshold
```

```python
df[df.price_per_sqft<min_threshold]
```

```python
df[df.price_per_sqft>max_threshold]
```

```python
df2=df[(df.price_per_sqft<max_threshold)&(df.price_per_sqft>min_threshold)]
df2        # Step 2
```

```python
df2.sample(10)
```

# Remove Outliers using 4 standard deviation  ,on step 2 dataframe

```python
df2.price.mean()
```

```python
df2.price.std()
```

```python
upper_limit=df.price.mean()+4*df2.price.std()
```

```python
lower_limit=df.price.mean()-4*df2.price.std()
```

```python
df2[(df.price>upper_limit)|(df.price<lower_limit)]
```

```python
df_no_outlier_std_dev=df2[(df2.price<upper_limit)|(df2.price>lower_limit)]
df_no_outlier_std_dev.shape
```

```python
df2.shape[0] - df_no_outlier_std_dev.shape[0]
```

```python
df_no_outlier_std_dev
```

# plot bell curve

```python
plt.hist(df_no_outlier_std_dev.price,bins=28,rwidth=0.8)
plt.xlabel('price')
plt.ylabel('Count')
plt.show()
```

```python
df_no_outlier_std_dev.price.min()
df_no_outlier_std_dev.price.max()
```

```python
from scipy.stats import norm
plt.hist(df_no_outlier_std_dev.price,bins=20,rwidth=0.8,density=True)
plt.xlabel('price')
plt.ylabel('count')
rng=np.arange(df_no_outlier_std_dev.price.min(),df_no_outlier_std_dev.price.max(),0.1)
plt.plot(rng,norm.pdf(rng,df_no_outlier_std_dev.price.mean(),df_no_outlier_std_dev.price.std()))
plt.show()
```

# Use Zscore of 4 to remove outliers

```python
df2['Zscore']=(df2.price-df2.price.mean())/df2.price.std()
df2
```

```python
df2[df2['Zscore']>4]
```

```python
df2[df2['Zscore']<4]
```

```python
df_no_outliers=df2[(df2.Zscore>-4) &(df2.Zscore<4)]
df_no_outliers
```
