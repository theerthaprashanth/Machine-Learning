

### üåº Imagine this:

You have **5 students** in a class.
They all got marks like this:

| Student | Marks |
| ------- | ----- |
| A       | 95    |
| B       | 90    |
| C       | 30    |
| D       | 35    |
| E       | 32    |

---

### Step 1Ô∏è‚É£ ‚Äî What we want to do

We want to **group students** based on how similar their marks are.
üëâ Students who have similar marks should be in the same group.

So here:

* A and B have high marks (close to each other).
* C, D, and E have low marks (close to each other).

---

### Step 2Ô∏è‚É£ ‚Äî What the computer does

The computer looks at **how close** or **how far** each pair of students is.

Then it groups them **step by step** like this:

1. First, it sees that **A** and **B** are most similar ‚Üí groups them together.
   (Group 1 ‚Üí {A, B})

2. Then, it sees that **C**, **D**, and **E** are close ‚Üí groups them together.
   (Group 2 ‚Üí {C, D, E})

3. Finally, it puts both groups together into one big group ‚Üí
   (All students together)

---

### Step 3Ô∏è‚É£ ‚Äî Why called ‚ÄúHierarchical‚Äù

Because it groups things **step by step**, like this tree:

```
             All students
             /          \
       (A,B)            (C,D,E)
```

So it‚Äôs like a **family tree of clusters** ‚Äî that‚Äôs what ‚Äúhierarchical‚Äù means.

---

### Step 4Ô∏è‚É£ ‚Äî Why it‚Äôs ‚Äúunsupervised‚Äù

Because:

* We didn‚Äôt tell the computer who should be in which group.
* It found the groups **on its own** by checking which marks are close.

That‚Äôs why it‚Äôs called **unsupervised learning**.

---

### üß† Simple summary:

| Term                        | Meaning                                           |
| --------------------------- | ------------------------------------------------- |
| **Clustering**              | Grouping similar things                           |
| **Hierarchical clustering** | Grouping step-by-step (like a tree)               |
| **Unsupervised**            | No labels given ‚Äî computer finds groups by itself |

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

x = [4, 5, 10, 4, 3, 11, 14 , 6, 10, 12]
y = [21, 19, 24, 17, 16, 25, 24, 22, 21, 21]

data = list(zip(x, y))

linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)

plt.show()
````
